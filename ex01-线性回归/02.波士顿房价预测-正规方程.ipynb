```python
"""
案例:
    演示 正规方程法 线性回归对象 完成 波士顿房价预测

回顾:
    线性回归算法 属于 有监督学习 有特征 有标签, 且标签是连续的.
    线性回归分类:
        一元线性回归: 1个特征列, 1个标签列
        多元线性回归: 多个特征列, 1个标签列
    如何衡量线性回归模型的好坏?
        思路:
            预测值和真实值之间的误差, 误差越小, 模型越好.  =>损失函数
        具体方案:
            1: 最小二乘法        每个(样本)误差平方和
            2: 均方误差(MSE)     每个(样本)误差平方和 / 样本总数
            3: 均方根误差(MSE)    sqrt(每个(样本)误差平方和 / 样本总数)
            4:平均绝对无擦(MAE)   abs(每个(样本)误差平方和 / 样本总数)

    如何让损失函数最小?
        思路1: 梯度下降法
            全梯度下降法(FGD)
            随机梯度下降法(SGD)
            小批量梯度下降法(BGD)
            随即平均梯度下降法(SAG)
        思路2: 正规方程法
"""

# 0.导包
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error
from sklearn.linear_model import Ridge, RidgeCV

url = 'https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv'

df = pd.read_csv(url)
# print(df)
# print(f"特征数据: {df.shape}")
# print(f"标签数据: {df['medv']}")

# 1.数据准备
x = df.drop('medv', axis=1)
y = df['medv']
# 2.数据分割
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=22)
# 3.数据预处理
# 3.1 创建标准化对象
transfer = StandardScaler()
# 3.2 对训练集进行标准化
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)

# 4.模型训练
# 4.1 创建 线性回归 正规方程 模型对象
estimator = LinearRegression(fit_intercept=True)
# 4.2 训练模型
estimator.fit(x_train, y_train)
# 4.3 打印权重和编制
print(f"模型的权重是: {estimator.coef_}")
print(f"模型的偏置是: {estimator.intercept_}")

# 5. 模型预测
y_pred = estimator.predict(x_test)
print(f"预测结果: {y_pred}")

# 6.模型评估
print(f"模型的均方误差(MSE)是: {mean_squared_error(y_test, y_pred)}")
print(f"模型的均方根误差(RMSE)是: {root_mean_squared_error(y_test, y_pred)}")
print(f"模型的平均绝对误差(MAE)是: {mean_absolute_error(y_test, y_pred)}")
```
