# 单变量线性回归

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('ex1data1.txt', header=None, names=['Population', 'Profit'])
data.head()
```

```python
data.describe()
```

```python
# 数据图示
"""
scatter:散点图
figsize=(12, 8):画布尺寸宽12英寸,高8英寸
"""
data.plot(kind='scatter', x='Population', y='Profit', figsize=(12, 8))
plt.show()
```
